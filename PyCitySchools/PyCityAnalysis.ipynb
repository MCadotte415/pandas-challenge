```python
#Library
import pandas as pd
import numpy as np
```

```python
#find files
school_data_file = "resources/schools_complete.csv"
student_data_file = "Resources/students_complete.csv"
```

```python
#read files
school_data = pd.read_csv(school_data_file)
student_data = pd.read_csv(student_data_file)
```

```python
#merge into one dataset
complete_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
complete_data.head()
```

```python
#total of unique schools
school_names = complete_data['school_name'].unique()
school_count = len(school_names)
print (school_count)
```

```python
#total number of students
student_count = school_data['size'].sum()
print(student_count)
```

```python
#the total budget
total_budget =school_data['budget'].sum()
print(total_budget)
```

```python
#average math score
avg_math_score= complete_data['math_score'].mean()
print(avg_math_score)
```

```python
#average reading score
avg_reading_score = complete_data['reading_score'].mean()
print(avg_reading_score)
```

```python
#percentage of students who passed math
passing_math_count = complete_data[(complete_data["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
print(passing_math_percentage)
```

```python
#percentage of students who passed reading
passing_reading_count = complete_data[(complete_data["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
print(passing_reading_percentage)
```

```python
#percentage of students that passed math and reading
passing_math_reading_count = complete_data[
    (complete_data["math_score"] >= 70) & (complete_data["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
print(overall_passing_rate)
```

```python
#summary snapshot
district_summary = pd.DataFrame({
    
    "Total Schools": [school_count],
    "Total Students": [student_count],
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math_score],
    "Average Reading Score": [avg_reading_score],
    "% Passing Math":[passing_math_percentage],
    "% Passing Reading": [passing_reading_percentage],
    "Overall Passing Rate": [overall_passing_rate]

})
#formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
#show the DataFrame
district_summary
```

```python
#code to select school type
school_types = school_data.set_index(["school_name"])["type"]
by_school = complete_data.set_index("school_name").groupby('school_name')
print(school_types)
```

```python
#total student count
per_school_count = by_school["Student ID"].count()
print(per_school_count)
```

```python
#total budget and capita spending
per_school_budget = complete_data.groupby(["school_name"]).mean()["budget"]
per_school_capita = per_school_budget / per_school_count
print(per_school_capita)
```

```python
#average test scores
avg_math = by_school['math_score'].mean()
avg_read = by_school['reading_score'].mean()
print(avg_math)
print(avg_read)
```

```python
#number of schools with math scores 70 ot higher
pass_math = complete_data[complete_data['math_score'] >= 70].groupby('school_name')['Student ID'].count()/per_school_count
```

```python
#number of schools with reading scores 70 or higher
pass_read = complete_data[complete_data['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/per_school_count
```

```python
#calculate the schools that passed both with scores of 70 or higher
pass_math_and_reading = complete_data[
    (complete_data["reading_score"] >= 70) & (complete_data["math_score"] >= 70)
]
(pass_math_and_reading).head()
```

```python
#code for passing rate
pass_math = complete_data[complete_data['math_score'] >= 70].groupby('school_name')['Student ID'].count()/per_school_count
pass_read = complete_data[complete_data['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/per_school_count
overall = complete_data[(complete_data['reading_score'] >= 70) & (complete_data['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/per_school_count 
```

```python
#DataFrame for School Summary
per_school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": per_school_count,
    "Per Student Budget": per_school_capita,
    "Total School Budget": per_school_budget,
    "Average Math Score": avg_math,
    "Average Reading Score": avg_read,
    '% Passing Math': pass_math,
    '% Passing Reading': pass_read,
    "Overall Passing Rate": overall
})
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)
 
per_school_summary
```
